#**STEP 1:Import Required Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Description**

•	pandas is used to read and process data.

•	matplotlib is used for plotting graphs.


#**STEP 3: Create CSV File**

```python
data_dict = {
    "Time": [0, 1, 2, 3, 4, 5],
    "Distance": [0, 12, 30, 55, 90, 130]
}

df = pd.DataFrame(data_dict)
df.to_csv("spot_speed_data.csv", index=False)
print("CSV file created successfully!")
```

#**STEP 4: Load CSV File**

```python
data = pd.read_csv("spot_speed_data.csv")
print("\nSpot and Speed Dataset:\n")
print(data)
```

**Description**

•	Reads CSV data into a DataFrame.

•	Displays the dataset for verification.


#**STEP 5: Calculate Speed**

```python
data['Speed'] = data['Distance'].diff() / data['Time'].diff()
data.fillna(0, inplace=True)
```

**Description**

•	Speed formula:

Speed = ΔDistance / ΔTime
      
•	.diff() calculates the difference between consecutive values.

•	First speed value is NaN, replaced with 0.


#**STEP 6: Perform Data Analysis**

```python
print("\n--- Speed Analysis ---")
print("Average Speed:", data['Speed'].mean(), "m/s")
print("Maximum Speed:", data['Speed'].max(), "m/s")
print("Minimum Speed:", data['Speed'].min(), "m/s")
```

#**Description**

•	Calculates average, maximum, and minimum speed.

•	Helps understand movement behavior.


#**STEP 7: Spot Analysis Graph (Distance vs Time)**

```python
plt.figure()
plt.plot(data['Time'], data['Distance'])
plt.xlabel("Time (seconds)")
plt.ylabel("Distance (meters)")
plt.title("Spot Analysis: Distance vs Time")
plt.grid(True)
plt.show()
```

#**Description**

•	Shows how distance changes with time.

•	Represents spot analysis.

#**STEP 8: Speed Analysis Graph (Speed vs Time)**

```python
plt.figure()
plt.plot(data['Time'], data['Speed'])
plt.xlabel("Time (seconds)")
plt.ylabel("Speed (m/s)")
plt.title("Speed Analysis: Speed vs Time")
plt.grid(True)
plt.show()
```

#**Description**

•	Visualizes speed variations over time.

•	Useful for identifying acceleration or deceleration.

#**STEP 8: Complete Code**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("spot_speed_data.csv")

# Calculate speed
data['Speed'] = data['Distance'].diff() / data['Time'].diff()
data.fillna(0, inplace=True)

# Display data
print("\nSpot and Speed Data:\n")
print(data)

# Speed analysis
print("\n--- Speed Analysis ---")
print("Average Speed:", data['Speed'].mean(), "m/s")
print("Maximum Speed:", data['Speed'].max(), "m/s")
print("Minimum Speed:", data['Speed'].min(), "m/s")

# Spot analysis graph
plt.figure()
plt.plot(data['Time'], data['Distance'])
plt.xlabel("Time (seconds)")
plt.ylabel("Distance (meters)")
plt.title("Spot Analysis (Distance vs Time)")
plt.grid(True)
plt.show()

# Speed analysis graph
plt.figure()
plt.plot(data['Time'], data['Speed'])
plt.xlabel("Time (seconds)")
plt.ylabel("Speed (m/s)")
plt.title("Speed Analysis (Speed vs Time)")
plt.grid(True)
plt.show()
```
